```python
from netCDF4 import Dataset
import pandas as pd
```

```python
ncfile = Dataset("rtg_sst_2017_03_13.nc")
pc     = pd.read_csv('/mnt/j/Gdrive/Tesis-Maestria/SALIDAS/csv/puntos_costa.csv')
pc1    = pc
```

```python
lon = pd.DataFrame({'lon':ncfile.variables['lon'][:]})
lon = lon-360
lat = pd.DataFrame({'lat':ncfile.variables['lat'][:]})
```

```python
for index, row in pc1.iterrows():
    #lat_i = row['Y']
    #lon_i = row['X']
    id_lat = lat.iloc[(lat['lat']-row['Y']).abs().argsort()[:1]].index
    id_lon = lon.iloc[(lon['lon']-row['X']).abs().argsort()[:1]].index
    tsm    = ncfile.variables['t'][0,id_lat,id_lon].item()-273.15
    pc1.loc[index,'cost']=tsm
    print(tsm)
    #print(str(row['Y'])+' == '+str(lat.loc[id_lat]['lat'].item()))
    #print(id_lon)
```
